### Figura 3D con Plotly express

```python
#Primero pasamos "local, empate, visita" como columnas
#pandas pivot: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.pivot.html

df_3d = df_league_winner.pivot(columns = "winner", values = "partidos", index = "league")
print(df_3d)
df_3d.reset_index(inplace = True)
df_3d
```

```python
figura_3d = px.scatter_3d(df_3d, x="local", y="visita", z="empate",text = "league")
figura_3d.update_traces(mode = "markers")
```

## Histogramas en Plotly

https://plotly.com/python/histograms/

https://plotly.github.io/plotly.py-docs/generated/plotly.graph_objects.Histogram.html

```python
#Creamos un numpy array random con 100 numeros entre el 1 y el 10
x1_random = np.random.randint(11, size=100)

#Creamos dos histogramas
fig1 = go.Figure(data=[go.Histogram(x= x1_random)])
fig2 = go.Figure(data=[go.Histogram(x= x1_random, histnorm='probability')])

fig1.show()
fig2.show()
#Plotly trabaja con una figura general, a la cual se le pueden agregar distintos graficos
```

#### Conocemos las variables de una figura en plotly

```python
x2_random = np.random.randint(11, size=100)

fig = go.Figure()
fig.add_trace(go.Histogram(x= x1_random, histnorm='probability', name="x1"))
fig.add_trace(go.Histogram(x= x2_random, histnorm='probability', name="x2"))

fig.update_layout(title = "Histograma", height = 500, width = 750)
```
